```python
import re
```

```python
#opening and reading the latest version of text corpus

document_text = open(r"C://Users/gurka/Desktop/final_clean.txt", 'r')
text_string = document_text.read()

#specifying a match pattern to find word frequency using the letters a-z with a length of 3-15 letters

pattern = r'\b[a-z]{7,7}\b'
match_pattern = re.findall(pattern, text_string, flags=re.IGNORECASE)

#Initializing an empty dictionary

frequency = { }
```

```python
#iterating through the file and adding key value {word:frequency} pairs to dictionairy

for word in match_pattern:
    count = frequency.get(word, 0)
    frequency[word] = count + 1

for word in frequency:
    print(word + ' => ' + str(frequency[word]))
```

```python
#sort the dictionairy in descending order of high to low word frequency

myList = sorted(frequency.items(), key=lambda x: x[1], reverse=False)

#taking a look at the results

myList
```
